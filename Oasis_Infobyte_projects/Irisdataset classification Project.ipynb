```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
import warnings 
from sklearn.preprocessing import LabelEncoder 
from sklearn.linear_model import LogisticRegression
```

## Data Read, Clean and Preprocessing

```python
df =pd.read_csv('/kaggle/input/iris-flower-dataset/IRIS.csv')
df.head(5)
```

```python
# Describing the dataset
df.describe()
```

```python
# Checking Dtypes, Nanvalues of the columns
df.info()
```

```python
# Checking for Nan values 
df.isnull().sum()
```

## Data Label Encoding 

```python
# Convesting target columns into binary data with label encoder
le = LabelEncoder()
df['species'] = le.fit_transform(df['species'])
```

```python
df.species.value_counts()
```

## Saprating Data Into X and Y 

```python
X= df.drop(['species'], axis=1)
y= df['species']
```

```python
#Checking the shape of the X and y
print(X.shape, y.shape)
```

## Model Building 

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
print(x_train.shape, x_test.shape,
     y_train.shape, y_test.shape)
```

```python
x_train
```

```python
# Model -1 Logstic Regression 
logr = LogisticRegression()
logr.fit(x_train, y_train)
```

```python
y_pred = logr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, y_pred))
```

```python
# Model -2 Decision Tree 
from sklearn.tree import DecisionTreeClassifier

Dtc = DecisionTreeClassifier()
Dtc.fit(x_train, y_train)
```

```python
y_pred_dtc = Dtc.predict(x_test)
y_pred_dtc
```

```python
print(classification_report(y_test, y_pred_dtc))
```
